```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

```
pd.plotting.parallel_coordinates(df,'Name');
```

```
pd.plotting.parallel_coordinates(df,'Name',color=('#F73333','#4ECDC4','#C7F464')); #html color
```

```
pd.plotting.parallel_coordinates(df,'Name',color=('r','g','b')); #ตัวย่อของสี R G B W C M Y K
```

```
pd.plotting.parallel_coordinates(df,'Name',color=((0.6,0,1),(0,1,0.5),(0,0.7,1))); #rgb value
```

#Parameter: colormap

```
pd.plotting.parallel_coordinates(df,'Name',colormap='cool'); #เป็นการกำหนดเฉดสีที่กำหนดมาแล้ว
```

```
pd.plotting.parallel_coordinates(df,'Name',colormap='winter');
```

#เริ่ม Matplotlib

```
from matplotlib import pyplot as plt
import numpy as np
```

#แปลง data จาก pandas dataframe เป็น numpy array

```
np_data = df.iloc[:,:-1].to_numpy()
np_data
```

#Visualize array data ด้วย pixel

```
import matplotlib
```

```
matplotlib.rcParams['figure.figsize']=[15,15] #กำหนดขนาดของ figure
plt.imshow(np_data[:50,:],interpolation='nearest'); #โชว์รูป
```

```
plt.imshow(np_data[50:100,:],interpolation='nearest');
```

#การใช้ subplot

```
plt.subplot(1,3,1) # subplot(มีกี่แถว,มีกี่หลัก,เป็นตัวที่เท่าไหร่)
plt.imshow(np_data[:50,:],interpolation='nearest',cmap='magma');
plt.subplot(1,3,2)
plt.imshow(np_data[50:100,:],interpolation='nearest',cmap='magma');
plt.subplot(1,3,3)
plt.imshow(np_data[100:150,:],interpolation='nearest',cmap='magma');
```

```
plt.imshow(np_data[:10,:],interpolation='nearest',cmap='magma',alpha=0.5);
```
